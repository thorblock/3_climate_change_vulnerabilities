```python
import pandas as pd
import matplotlib.pyplot as plt 
import json
```

```python
# Load into a DataFrame named 'pdx_heat'
pdx_heat = pd.read_csv('portland_heat.csv')

pdx_heat
```

```python
zip97210 = pdx_heat[pdx_heat['OBJECTID'].isin([1915, 1918, 1965, 1908])]
zip97233 = pdx_heat[pdx_heat['OBJECTID'].isin([1970, 1969, 1976, 1977])]
zip97229 = pdx_heat[pdx_heat['OBJECTID'].isin([1942])]


```

```python

am97210 = float(zip97210[['am']].mean())
af97210 = float(zip97210[['af']].mean())
pm97210 = float(zip97210[['pm']].mean())
anam97210 = float(zip97210[['anomaly_am']].mean())
anaf97210 = float(zip97210[['anomaly_af']].mean())
anpm97210 = float(zip97210[['anomaly_pm']].mean())



real_am97210 = am97210 + anam97210
real_af97210 = af97210 + anaf97210
real_pm97210 = pm97210 + anpm97210


real_pm97210
```

```python

am97229 = float(zip97229[['am' ]].mean())
af97229 = float(zip97229[['af' ]].mean())
pm97229 = float(zip97229[['pm' ]].mean())
anam97229 = float(zip97229[['anomaly_am' ]].mean())
anaf97229 = float(zip97229[['anomaly_af' ]].mean())
anpm97229 = float(zip97229[['anomaly_pm' ]].mean())

real_am97229 = am97229 + anam97229
real_af97229 = af97229 + anaf97229
real_pm97229 = pm97229 + anpm97229



#print(f"{am97229}, {anam97229}, {af97229}, {anaf97229}, {pm97229}, {anpm97229}")
```

```python
am97233 = float(zip97233[['am' ]].mean())
af97233 = float(zip97233[['af' ]].mean())
pm97233 = float(zip97233[['pm' ]].mean())
anam97233 = float(zip97233[['anomaly_am' ]].mean())
anaf97233 = float(zip97233[['anomaly_af' ]].mean())
anpm97233 = float(zip97233[['anomaly_pm' ]].mean())

real_am97233 = am97233 + anam97233
real_af97233 = af97233 + anaf97233
real_pm97233 = pm97233 + anpm97233
```

```python
canopy97210 = float(zip97210[['mean_canopy' ]].mean())
canopy97229 = float(zip97229[['mean_canopy' ]].mean())
canopy97233 = float(zip97233[['mean_canopy' ]].mean())

print(f"{canopy97233}, {canopy97229}, {canopy97210}")
```

```python
data = [    ["Morning Air temperature",[am97229],[am97210],[am97233]],["Morning Ground temperature",[real_am97229],[real_am97210],[real_am97233]],["Afternoon Air temperature",[af97229],[af97210],[af97233]],["Afternoon Ground temperature",[real_af97229],[real_af97210],[real_af97233]],["Evening Air temperature",[pm97229],[pm97210],[pm97233]],["Evening Ground temperature",[real_pm97229],[real_pm97210],[real_pm97233]],["Canopy", [canopy97229],[canopy97210],[canopy97233]]]

df = pd.DataFrame(data, columns=['Time', '97229', '97210','97233'])

# Convert DataFrame to JSON
json_data = df.to_json(orient='split')

# Define the file path where you want to save the JSON file
json_file_path = 'by_zip.json'

# Write the JSON data to the file
with open(json_file_path, 'w') as json_file:
    json_file.write(json_data)
```
